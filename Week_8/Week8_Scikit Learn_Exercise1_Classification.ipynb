```python
from sklearn import svm
X = [[0, 0], [1, 1]]
y = [0, 1]
clf = svm.SVC()
clf.fit(X, y)
```

```python
clf.predict([[2., 2.]])
```

```python
clf.support_vectors_
clf.support_
clf.n_support_
```

```python
X = [[0], [1], [2], [3]]
Y = [0, 1, 2, 3]
clf = svm.SVC(decision_function_shape='ovo')
clf.fit(X, Y)

dec = clf.decision_function([[1]])
dec.shape[1]

clf.decision_function_shape = "ovr"
dec = clf.decision_function([[1]])
dec.shape[1]
```

```python
lin_clf = svm.LinearSVC()
lin_clf.fit(X, Y)
dec = lin_clf.decision_function([[1]])
dec.shape[1]
```
